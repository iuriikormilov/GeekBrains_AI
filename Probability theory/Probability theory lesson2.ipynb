__Задача 1__

Контрольная работа состоит из пяти вопросов. На каждый вопрос приведено четыре варианта ответа, один из которых правильный. Составьте закон распределения числа правильных ответов при простом угадывании. Найдите `M(X)`, `D(X)`.

```python
import numpy as np

def combinations(n: int, k: int) -> int:
    """Число сочетаний.
    """
    
    return np.math.factorial(n) // (np.math.factorial(k) * np.math.factorial(n - k))

def permutations(n: int) -> int:
    """Число перестановок.
    """
    
    return np.math.factorial(n)

def arrangements(n: int, k: int) -> int:
    """Число размещений.
    """
    
    return np.math.factorial(n) // np.math.factorial(n - k)
```

__Решение__

Задача на использование биномиального закона распределения.

Вариантов выбора ответа 4, следовательно вероятность выбора правильного ответа p = 0.25. Вопросов 5, n = 5. k - количество верно отгаданных вопросов может быть от 0 до 5.

$$P(X = k) = C_n^k p^k (1 - p)^{n - k}$$

```python
for k in range(0, 6):
    n = 5
    p = 0.25
    q = 1 - p
    P = combinations(n, k)*(p**k)*(q**(n-k))
    print(f"Вероятность ответить на {k} вопросов: {P}")
```

Посчитаем Матожидание и дисперсию: 
    $$M(X) = np, \: D(X) = np(1 - p).$$

```python
M = n*p
D = n*p*q
print(f"Матожидание равно: {M}\nДисперсия равна: {D}")
```

__Задача 2__

Вероятность попадания в цель при одном выстреле равна 0.01. Какова вероятность того, что число попаданий при 200 выстрелах будет не менее 5 и не более 10?

__Решение__

Мы можем допустить, что рассматриваемая величина имеет распределение Пуассона, так как количество выстрелов довольно высоко, а вероятность мала:
$$P(X = k) = \dfrac{a^k e ^{-a}}{k!}.$$
Так как нам надо найти вероятность попаданий в диапазоне, просуммируем вероятности от 5 до 10 включительно.

```python
P = 0
for k in range(5, 11):
    n = 200
    p = 0.01
    a = n * p
    Pi = (a**k*np.math.exp(-a))/(np.math.factorial(k))
    P += Pi
print(f"Вероятность того, что число попаданий будет не менее 5 и не более 10: {P}")
```

__Задача 3__

В условиях предыдущей задачи сколько выстрелов нужно сделать чтобы быть уверенным с вероятностью 0.9, что хотя бы 10 раз будет совершено попадание?

_Подсказка_. Как посчитать `P(X >= k)` для величины `X`, имеющей распределение Пуассона? Здесь предстоит немножко покодить.

```python
Psum = 1
n = 10
while Psum > 0.1:
    Psum = 0
    for k in range(0, 10):
        p = 0.01
        a = n * p
        P = (a**k*np.math.exp(-a))/(np.math.factorial(k))
        Psum += P
        # print(f"Вероятность попасть {k} раз в мишень: {P}")
    n += 1
print(f"Необходимо сделать не менее {n - 1} выстрелов. Вероятность попасть в мишень от 0 до 9 раз в таком случае составляет {Psum}")
print(f"Следовательно с вероятностью {1 - Psum} можно сказать, что хотя бы 10 из {n - 1} попадут в цель")
```
