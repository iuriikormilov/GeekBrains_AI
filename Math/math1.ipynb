Постройте на одном графике две кривые y(x) для функции двух переменной y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)            
y1 = np.cos(1*x) 
y2 = np.cos(2*x)  

fig = plt.figure()
ax1 = fig.add_subplot(111)

line1, = ax1.plot(x, y1)
ax1.set_xlabel('x')
ax1.set_ylabel('y')

line2, = ax1.plot(x, y2, color='green')
plt.title("График функции y = cos(kx) при разных k")
plt.legend((line1, line2), ('k = 1', 'k = 2'))
plt.show()
```
