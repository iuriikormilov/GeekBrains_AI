<p>Напишите код, который будет переводить полярные координаты в декартовы.</p>

```python
def polar_to_decart(r, phi):
    x = r * np.cos(phi)
    y = r * np.sin(phi)
    return x, y
```

Напишите код, который будет рисовать график окружности в полярных координатах.

<h3>4. Задание (в программе)</h3>
<p>Решите систему уравнений:</p>
<p>y = x2 – 1</p>
<p>exp(x) + x∙(1 – y) = 1</p>

```python
fig= plt.figure(figsize=(6,6))

x = np.linspace(-2, 5, 500)
y1 = x**2 - 1
y2 = (-1 + np.exp(x) + x)/x
plt.plot(x, y1)
plt.plot(x, y2)

plt.xlim(-2, 5) 
plt.ylim(-1, 25)
idx = np.argwhere(np.diff(np.sign(y1 - y2))).flatten()
plt.plot(x[idx], y1[idx], 'bo')

plt.grid()
plt.show()
```

<p>Решите систему уравнений и неравенств:</p>
<p>y = x2 – 1</p>
<p>exp(x) + x∙(1 – y) > 1</p>

```python
from sympy import var, plot_implicit, Eq
var('x y')
plot_implicit((sy.exp(x)+x*(1-y))>1, (x, -2, 5), (y, -1, 25), line_color="yellow");
plot_implicit(Eq(y - x**2, -1),(x, -2, 5), (y, -1, 25));
```
